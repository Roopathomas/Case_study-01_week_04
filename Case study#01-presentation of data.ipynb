# 1. Read the dataset to the python environment.

```python
# importing pandas as pd
import pandas as pd

#Read dataset 
Data=pd.read_csv(r"C:\Users\Justin Thomas\Downloads\cars_data.csv")
Data.head()
```

# 2. Check for the null values present in the dataset.


```python
#null values present in the dataset

Data. isnull()
```

# Result

# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
# Implementation of matplotlib function
import matplotlib.pyplot as plt

plt.figure()

#plotting bar graph 
plt.bar(Data["Buyer Gender"],height=0.10)

#plotting title
plt.title("Male vs Female buyers participated in the sales",fontsize=16)

#plotting x axis
plt.xlabel("Gender")

#plotting y axis
plt.ylabel("Frequency")
plt.show()
```

# Result
**Above Graph shows that, equal Male and Female buyers participated in the sales**

# 4. Find the top 5 cars based on their sales price.

```python
## five largest values in column sales price

Data.nlargest(5, ['Sale Price'])
```

# Result
**Volkswagen-Jetta III, Audi-A6, Lexus-LX, Mazda-MX-6 and Lexus-ES are the top 5 cars based on their sales price.**

# 5. Find the least 5 cars based on their Resell price.

```python
## five least values in column Resell price

Data.nsmallest(5,['Resell Price'])
```

# Result
**Chevrolet-1500, GMC-Savana 1500, Jangur-XF, Mitsubishi-3000GT and Infiniti-G37 are the least 5 cars based on their Resell price.**
